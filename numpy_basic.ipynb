```python
import math
```

```python
customers =[[17,750],[18,800],[19,900],[3,400]]
new_customer = [3,400]

distance = []
for customer in customers:
    dist = math.sqrt((customer[0] - new_customer[0])**2 + (customer[1] - new_customer[1])**2)
    distance.append(dist)


# for nd
for i in range(len(customers)):
    dist = math.sqrt((customers[i][0] - new_customer[0])**2 + (customers[i][1] - new_customer[1])**2)
    distance.append(dist)
```

```python
def calculate_distance(customers, new_customer):
    new_distance = []
    for i in range(len(customers)):
        dist = math.sqrt((customers[i][0] - new_customer[0])**2 + (customers[i][1] - new_customer[1])**2)
        new_distance.append(dist)
    return new_distance


```
