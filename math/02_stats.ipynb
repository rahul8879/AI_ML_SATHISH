```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy 
```

```python
delivery_time_in_minutes = np.random.normal(loc=30, scale=5, size=1000)
plt.hist(delivery_time_in_minutes, bins=30, density=True, alpha=0.5)
plt.show()


# for confidence interval
confidence_interval = scipy.stats.norm.interval(0.95, loc=np.mean(delivery_time_in_minutes), scale=scipy.stats.sem(delivery_time_in_minutes))
```
