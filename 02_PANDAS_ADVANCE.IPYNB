```python
data.rename(columns={'Description': 'Product_Description'}, inplace=True)
```

```python
def data_checks(df):
    checks ={
        'size_of_data': df.shape[0],
        'num_of_columns': df.shape[1],
        'missing_values': df.isnull().sum().sum(),
        'pct_missing': round(df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100,2)
    }
    return checks

data_checks(data)
```

```python
def tagg_to_total_sales(sales):
    if sales > 100:
        return "High"
    elif sales > 50:
        return "Medium"
    else:
        return "Low"
    

data['sales_tag'] = data['total_price'].apply(tagg_to_total_sales)
```
